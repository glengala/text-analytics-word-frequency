```python
# Let the user enter the file:
policy=input("Enter the policy text file:")
sightword=input("Enter the sight text file:")

# Open the above files:
fname=open(policy)
fsword=open(sightword)

# Create an empty list to contain all the sight words
sword=[]

# Create an empty dictionary for the word and its counts
dict_words=dict()

# Create an empty list to contain words extracted from the policy
words=[]

# Get the list of words from the policy line by line
for line in fname:
    line=line.rstrip()
    ln= line.split()
    for i in ln:
        words.append(i.lower())
        
# Add sight words to sword list created above
for sw in fsword:
    sw=sw.rstrip()
    s= sw.split()
    s= sw.split(",")
    for i in s:
        if len(i)>0:
            i=i.strip()
            sword.append(i.lower())
            
words_filtered=[x for x in words if x not in sword]

# Remove plural form of word as it doesn't contain any meaningful interpretation

words_sg=[w[:-1] if w.endswith("s") else w for w in words_filtered]  
    
# Add word and its count into the dictionary dict_words
for word in words_sg: 
    dict_words[word]=dict_words.get(word,0)+1
sort_dict=sorted(dict_words.items(), key=lambda x: x[1], reverse=True)
#print(f"After remove sight words: {sort_dict}")

# Print top 20 most frequency words
top_n=20
print(f"___Top {top_n} most frequency words are:")
for key,value in sort_dict[:top_n]:
    print(key,value)

```
